```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import datetime as dt
from datetime import datetime as dtdt
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import ticker
import matplotlib.dates as mdates
from matplotlib.dates import date2num
from matplotlib.dates import DateFormatter
import seaborn as sns
sns.set()
```

```python
path_cs = "/Users/iomacbookair2/Documents/lab/DEIM2023/sentense_similarity/cosine_similarity_221212_ann_mon.csv"
df_cs = pd.read_csv((path_cs))
df_cs = df_cs.drop("texts_by_minute", axis=1)

path_count = "/Users/iomacbookair2/Documents/lab/DEIM2023/count/count_221212_ann_mon.csv"
df_count = pd.read_csv((path_count))
df_cs["count"] = df_count["counts"]
df = df_cs
df
# df.to_csv('cs_count_221212_ann_mon.csv', index=False)
```

```python
# Create a figure and an Axes object
fig, ax = plt.subplots(figsize=(25, 10))

ax.set_title("Average Similarity Scores and Tweet Counts of 221212_ann_mon")

ax.set_xlabel("Time", fontsize=20)
# Set the x-axis labels and rotate them by 90 degrees
ax.set_xticklabels(df["created_at"], rotation=90)
ax.set_xlim(-0.5, 124.5) # グラフの横幅と全体の端を揃える

# Set the y-axis label for the count data
ax.set_ylabel("Count", fontsize=20)

# Plot the count data as partially transparent vertical bars
ax.bar(df["created_at"], df["count"], color="#70d6ff", label="Count")

# Create a twin Axes object for the count data
ax2 = ax.twinx()

# Set the y-axis labels
ax2.set_ylabel("Average Similarity", fontsize=20)

# Plot the average similarity data
ax2.plot(df["created_at"], df["average_similarity"], color="red", label="Average Similarity")

# Adjust the position of the legend
ax.legend(loc="upper right", bbox_to_anchor=(1,0.98))
ax2.legend(loc="upper right", bbox_to_anchor=(1,0.94))

# Show the plot
plt.show()
```
